```python
import yfinance as yf
import pandas as pd
from pandas.tseries.offsets import MonthEnd  
```

```python
tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0].Symbol
```

```python
type(tickers)
```

```python
df_ = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[1]
```

```python
df_.Date = pd.to_datetime(df_.Date.Date)
```

```python
df_ = df_[df_.Date.Date >= '2020-01-01']
```

```python
tickers[~(tickers.isin(df_.Added.Ticker))]
```

```python
tickers_rem = df_.Removed.Ticker
```

```python
tickers = tickers._append(tickers_rem)
```

```python
tickers.drop_duplicates(inplace = True)
tickers.dropna(inplace = True)
```

```python
tickers
```

```python
start = '2020-01-01'
prices = []
symbols = []
```

```python
for symbol in tickers:
    df = yf.download(symbol,start=start)['Adj Close']
    if not df.empty:
        prices.append(df)
        symbols.append(symbol)
```

```python
allprices = pd.concat(prices,axis=1)
```

```python
allprices.columns = symbols 
```

```python
allprices.pct_change()
```

```python
(allprices.pct_change()+1).prod()-1
```

```python
all_monthly_return = allprices.pct_change().resample('M').agg(lambda x:(x+1).prod()-1)
```

```python
all_monthly_return
```

```python
allprices.resample('M').last()
```

```python
all_monthly_return_12 = all_monthly_return.rolling(12).agg(lambda x: (x+1).prod()-1)
```

```python
all_monthly_return_12.dropna(inplace=True)
```

```python
curr = all_monthly_return_12.iloc[0]
```

```python
win = curr.nlargest(10)
```

```python
win.index
```

```python
win_ret = all_monthly_return.loc[win.name + MonthEnd(1), win.index]
```

```python
win_ret
```

```python
win_ret.mean()
```

```python
def mom(all_monthly_return, lookback):
    all_monthly_ret_lb = all_monthly_return.rolling(lookback).agg(lambda x: (x+1).prod()-1)
    all_monthly_ret_lb.dropna(inplace=True)
    rets = []
    for row in range(len(all_monthly_ret_lb)-1):
        curr = all_monthly_ret_lb.iloc[row]
        win = curr.nlargest(10)
        win_ret = all_monthly_return.loc[win.name+MonthEnd(1),win.index]
        rets.append(win_ret.mean())
    return (pd.Series(rets)+1).prod()-1
    
```

```python
for lookback in range(1,13):
    print(mom(all_monthly_return, lookback))
```
